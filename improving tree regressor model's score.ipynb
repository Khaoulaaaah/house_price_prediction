```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, PowerTransformer

df0 = pd.read_csv("hf://datasets/daniilak/Russia_Real_Estate_2018_2021/data.csv", nrows=100000)
df0 = df0.drop(['date', 'time'], axis=1) 
df0 = df0.dropna() 
df0 = df0.drop_duplicates()  
df0 = df0[df0['price'] > 0] 
df0['price'] = np.log1p(df0['price']) 
df0['price'] = np.log1p(df0['price'])
non_numeric_cols = df0.select_dtypes(include=['object']).columns
df0 = pd.get_dummies(df0, columns=non_numeric_cols, drop_first=True)
x = df0.drop(['price'], axis=1)
y = df0['price']
x = pd.DataFrame(StandardScaler().fit_transform(x), columns=x.columns)  # Standardizing features
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
param_grid = {
    'n_estimators': [100],
    'max_depth': [20],
    'min_samples_split': [7],
    'min_samples_leaf': [4],
    'bootstrap': [True, False]
}
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)
best_rf = grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)
train_score = best_rf.score(X_train, y_train)
test_score = best_rf.score(X_test, y_test)
mse = mean_squared_error(np.expm1(y_test), np.expm1(y_pred_rf))  # Transform back from log scale
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred_rf))  # Transform back from log scale
print("Best Parameters:", grid_search.best_params_)
print("Training Score (R^2):", train_score)
print("Test Score (R^2):", test_score)
print("Mean Squared Error (MSE):", mse)
print("R-Squared (Test Set):", r2)
```

```python
def get_user_input():
    """
    Interactively asks the user for input values for the required and optional fields.
    
    Returns:
    - Dictionary containing the user's input for all fields.
    """
    print("Provide the following details to predict the house/real estate price.")
    print("(Press Enter to skip optional fields.)\n")
    
    # Required fields
    required_fields = ['geo_lat', 'geo_lon', 'region', 'level', 'levels', 'rooms', 'area']
    user_input = {}
    for field in required_fields:
        while True:
            value = input(f"Enter {field} (required): ")
            if value.strip():  # Ensure required fields are not empty
                try:
                    user_input[field] = float(value) if field not in ['region', 'rooms'] else int(value)
                    break
                except ValueError:
                    print(f"Invalid value for {field}. Please enter a number.")
            else:
                print(f"{field} is required. Please provide a value.")

    # Optional fields
    optional_fields = ['building_type', 'kitchen_area']
    for field in optional_fields:
        value = input(f"Enter {field} (optional): ")
        if value.strip():  # Allow skipping optional fields
            try:
                user_input[field] = float(value) if field == 'kitchen_area' else int(value)
            except ValueError:
                print(f"Invalid value for {field}. Skipping this field.")
    
    return user_input


def predict_price_custom(rf_model):
    """
    Predict the price of a house/real estate based on user-provided input.
    
    Args:
    - rf_model: Trained RandomForestRegressor model.
    
    Returns:
    - Predicted price (float)
    """
    # Get user input
    user_input = get_user_input()
    
    # Default values for optional fields
    defaults = {'building_type': 0, 'kitchen_area': 0}
    
    # Prepare input data with defaults for missing fields
    data = {field: user_input.get(field, defaults.get(field)) for field in 
            ['geo_lat', 'geo_lon', 'region', 'building_type', 'level', 'levels', 'rooms', 'area', 'kitchen_area']}
    
    # Convert to DataFrame
    input_df = pd.DataFrame([data])
    
    # Predict the price
    predicted_price = rf_model.predict(input_df)[0]
    print(f"\nPredicted price: {predicted_price:.2f} RUB")


# Example usage
predict_price_custom(rf_model)
```
